```python

def all_subsets(Vars,Contains=None,MinL=1,MaxL=None):
#       ** Find all posible subset of input varriables within range provided**
#       ** Contains limits to subsets including given variables**
    if MaxL == None:
        MaxL = len(Vars)
    Model = {}
    Cmax = 0
    c=0
    Model = {}
    for x in range(MinL,MaxL+1):
        Model[c]={}
        for mod in combinations(Vars, x):
            mod = list(mod)
            if Contains != None:
                if set(Contains).issubset(mod) == True:
                    Model[c]=[]
                    Model[c].append(mod)
                    c+=1

            else:
                Model[c]=[]
                Model[c].append(mod)
                c+=1
        
    P_m = np.zeros(shape=(len(Model)))
    P_m[:]=1/(c)
    Mi = np.arange(0,c)
    print('# of Models: ',c)
#     print(Cmax)
#     P_m = np.zeros(shape = (len(Model)+MinL,Cmax))
#     for M in Model:
#         for i in Model[M]:
#             P_m[M,i]=1/(p_t)
#     print(M2[1],Ml)
    return(Model,Mi,P_m,c)
```

```python
class ReadStandardTime:
    def __init__(self,Path):
        self.Master = pd.read_csv(Path,delimiter = ',',header = 0,na_values = -9999)
        self.Master = self.Master.set_index(pd.DatetimeIndex(pd.to_datetime(self.Master['datetime'])))
        self.Master['DOY'] = self.Master.index.dayofyear*1.0
        self.Master['HR'] = self.Master.index.hour*1.0
        
    def Scale(self,y_var,X_vars):
        self.y_var = y_var
        self.Data = self.Master[np.isfinite(self.Master[y_var])]
        self.Data = self.Data.interpolate().bfill()
        self.Data = self.Data.interpolate().ffill()
        self.y = self.Data[y_var].values
        YStandard = StandardScaler()
        self.YScaled = YStandard.fit(self.y.reshape(-1, 1))
        Yscale = self.YScaled.transform(self.y.reshape(-1, 1))
        self.y = np.ndarray.flatten(Yscale)
        self.Ytru = self.YScaled.inverse_transform(self.y.reshape(-1,1))
        X = self.Data[X_vars]
        self.input_shape = len(X_vars)
        XStandard = StandardScaler()
        self.XScaled= XStandard.fit(X)
        self.X = self.XScaled.transform(X)
        Filling = self.Master[X_vars]
        Filling = Filling.interpolate().bfill()
        Filling = Filling.interpolate().ffill()
        XStandard = StandardScaler()
        self.XFillScaled= XStandard.fit(Filling)
        self.X_fill = self.XScaled.transform(Filling)
        
    def TimeShape(self,rolls):
        X1 = self.X
        Xf = self.X_fill
        self.X_time = np.zeros(shape = (X1.shape[0],rolls+1,X1.shape[1]))
        self.X_time[:,0,:] = X1
        self.X_ftime = np.zeros(shape = (Xf.shape[0],rolls+1,Xf.shape[1]))
        self.X_ftime[:,0,:] = Xf
        if rolls > 0:
            for roll in range(0,rolls):
                X2 = np.roll(X1,(roll+1),axis=0)
                X2f = np.roll(Xf,(roll+1),axis=0)
                self.X_time[:,roll+1,:] = X2
                self.X_ftime[:,roll+1,:] = Xf
        self.X_time = self.X_time[rolls+1:,:,:]
        self.X_ftime = self.X_ftime[rolls+1:,:,:]
        self.y_time = self.y[rolls+1:]
        self.y_ftime = self.y[rolls+1:]
        self.TimeSteps = rolls+1
        
    def Fill(self,Y_Pred):
        Y_fill = self.YScaled.inverse_transform(Y_Pred.reshape(-1,1))
        nanz = np.zeros(shape=(self.TimeSteps,1))
        nanz[:,:] = np.nan
        Y_Pred = np.concatenate((nanz,Y_fill),axis=0).reshape(-1,1)
        self.Master['TempFIll'] = Y_Pred
        self.Master[self.y_var+'_Filled'] = self.Master[self.y_var].fillna(self.Master['TempFIll'])
```

```python
def LSTM_Model(Neurons,batch_size,time_steps,inputs,lr=1e-4,Memory=.9):
    import keras
    from keras.models import Sequential
    from keras.layers import Dense
    from keras.layers import LSTM
    from keras.wrappers.scikit_learn import KerasRegressor
    from keras.callbacks import EarlyStopping,ModelCheckpoint,LearningRateScheduler
    from keras import backend as K
    import tensorflow as tf
    config = tf.ConfigProto()
    config.gpu_options.per_process_gpu_memory_fraction = Memory
    session = tf.Session(config=config)
    model = Sequential()
    model.add(LSTM(Neurons, input_shape=(time_steps,inputs),stateful = False))
    model.add(Dense(1))
    NUM_GPU = 1 # or the number of GPUs available on your machine
    
    rmsp = keras.optimizers.RMSprop(lr = lr)
    gpu_list = []
    for i in range(NUM_GPU): gpu_list.append('gpu(%d)' % i)
    model.compile(loss='mean_squared_error', optimizer='adam')#,context=gpu_list) # - Add if using MXNET
    return(model)

def Train_Steps(Neurons,X_train,X_test,X_val,y_train,y_test,y_val,rs,i=None,Memory=None):
    epochs = 1000
    Scorez=[]
    lr = 1e-4
    Mod1 = LSTM_Model(Neurons,X_train.shape[0],X_train.shape[1],X_train.shape[2],lr=lr,Memory=Memory)
    killscore=0
    killmax = 3
    e = 0
    udate = 3
    np.random.seed(rs)
    while killscore < killmax and e < epochs:
        Mod1.fit(X_train,y_train,batch_size=X_train.shape[0], nb_epoch=1,shuffle=True,verbose=0)
        old_weights = Mod1.get_weights()
        Y = Mod1.predict(X_test,batch_size =X_test.shape[0])
        score = metrics.mean_squared_error(y_test,Y)
        Scorez.append(score)
        if e == 0:
            score_min=score
            min_weights=old_weights
        elif score < score_min:
            score_min = score
            min_weights=old_weights
            killscore = 0
        else:
            killscore +=1
        Mod1.reset_states()
        e +=1
    Mod1.set_weights(min_weights)
    Yval = Mod1.predict(X_val,batch_size = X_val.shape[0])
    MSE = (metrics.mean_squared_error(y_val,Yval))
    print(i,' MSE ', MSE)
    Scorez=np.asanyarray(Scorez)
    return(MSE,min_weights)
           
def RunFunc(i,processes,Modz,init):
#     print('kitty')
    Path = 'QuickData.csv' 
#     CH4_Model = ['Sedge','Temp','VWC','ustar','air_pressure','PPFD_Avg']
    RST = ReadStandardTime(Path)
    offset = 3/processes
    Memory = (math.floor(100/processes)- offset) * .01
    MSE = []
# #     List = [4,5,8,9]
# #     Model = [['Sedge','Temp','VWC','ustar','Shrubby'],
# # #              ['Sedge','Temp','VWC','ustar','AirTC_Avg'],
# # #              ['Sedge','Temp','VWC','ustar','PPFD_Avg'],
# # #              ['Sedge','Temp','VWC','ustar','air_pressure'],
# #              ['Sedge','Temp','VWC','ustar','PPFD_Avg','AirTC_Avg'],
# # #              ['Sedge','Temp','VWC','ustar','air_pressure','PPFD_Avg','AirTC_Avg'],
# # #              ['Sedge','Temp','VWC','ustar','air_pressure','PPFD_Avg','AirTC_Avg','RH'],
# #              ['Sedge','Temp','VWC','ustar','air_pressure','PPFD_Avg','AirTC_Avg','NR_Wm2_Avg'],
# #              ['Sedge','Temp','VWC','ustar','air_pressure','PPFD_Avg','AirTC_Avg','RH','NR_Wm2_Avg']]
# #     for M in Modz:
# #         Modz[str(StartSize)][M]
    M = Modz[1]
    print(M)
    RST.Scale('ch4_flux',M)
    RST.TimeShape(12)
    y = RST.y_time*1.0
    X = RST.X_time*1.0
    X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.1, random_state=init)
    X_train,X_val,y_train,y_val=train_test_split(X_train,y_train, test_size=0.11, random_state=init)
    mse,w=Train_Steps(15,X_train,X_test,X_val,y_train,y_test,y_val,rs=i,i=init,Memory=Memory)
#         weights[str(i)]=w
    MSE.append(mse)
# #     Neurons = [5,10,15,20,25,30]
# #     for N in Neurons:
# #         mse,w=Train_Steps(N,X_train,X_test,X_val,y_train,y_test,y_val,rs=i,i=i,Memory=Memory)
# # #         weights[str(i)]=w
# #         MSE.append(mse)
#     mse = np.asanyarray(MSE)
# #     print(MSE)
#     mse = 10
    return(mse)
```
